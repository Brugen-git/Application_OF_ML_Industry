```python
# Getting the attributes
attr = np.array(df)[:,:-1]
print("n The attributes are: ",attr)
```

```python
# Getting the values from last class
val = np.array(df)[:,-1]
print(val)
```

```python
# applying the find_s algorithm
def find_s(con, tar):
    yes_list=['YES','yes','y','Y']
    for i, val in enumerate(tar):
        if val in yes_list:
            specific_h = con[i].copy()
            break
            
    for i, val in enumerate(con):
        if tar[i] in yes_list:
            for x in range(len(specific_h)):
                if val[x] != specific_h[x]:
                    specific_h[x] = '?'
                else:
                    pass
    return specific_h
```

```python
# Printing the final hypothesis
print(find_s(attr, val))
```
